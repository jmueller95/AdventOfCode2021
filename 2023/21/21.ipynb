```python
import numpy as np
```

```python
input = "input.txt"
with open(input, 'r') as infile:
    #Today we split into characters
    data = np.array([[c for c in l.strip()] for l in infile.readlines()])
```

```python
data
```

Create distance map:

```python
distance_map = np.full_like(data, -1, dtype=int)
```

Find start coordinates:

```python
for i in range(len(data)):
    for j in range(len(data[0])):
        if data[i,j] == 'S':
            break
    if data[i,j] == 'S':
        break
```

```python
plot_list = [(i,j)]

counter = 0
while counter <= 64:
    new_plot_list = set()
    for x,y in plot_list:
        distance_map[x,y] = counter
        for i,j in [(x-1,y),(x,y-1),(x+1,y),(x,y+1)]:
                if 0 <= i < len(data) and 0 <= j < len(data[0]) and distance_map[i,j] == -1 and data[i,j] == '.':
                    new_plot_list.add((i,j))
    plot_list = new_plot_list
    counter += 1
```

```python
distance_map
```

```python
(distance_map % 2 == 0).sum()
```

```python
distance_map
```

It's modulo two because I can go back and forth. So anything I can reach after an even number of steps I can also reach after 64 steps. And if I can only reach it after an odd number of steps then it's impossible to go there in 64.

I expect that after a while, the growth rate should stabilize. So let's monitor the growth rate for some time:

```python
for si in range(len(data)):
    for sj in range(len(data[0])):
        if data[si,sj] == 'S':
            break
    if data[si,sj] == 'S':
        data[si,sj] =  '.'
        break
```

We don't use the distance map this time. Instead just instantly record the number of plots that can be reached.

```python
plot_list = [(si,sj)]
plots_per_step = dict()
counter = 0
while counter < 100:
    new_plots = set()
    for x,y in plot_list:
        for i,j in [(x-1,y),(x,y-1),(x+1,y),(x,y+1)]:
            if data[i%len(data),j%len(data[0])] == '.':
                new_plots.add((i,j))
    #plot_list = list(set((i%len(data), j%len(data[0])) for i,j in new_plots))
    plot_list = list(new_plots)
    counter += 1
    plots_per_step[counter] = len(new_plots)
```
