PART I

```python
input = "input.txt"
with open(input, 'r') as infile:
    data = [l.strip() for l in infile.readlines()]
```

```python
def check_arrangement(springs, numbers):
    res = []
    counter = 0
    for spring in springs:
        if spring == '#':
            counter += 1
        elif counter > 0:
            res.append(counter)
            counter = 0
    if counter > 0:
        res.append(counter)
    return res == numbers

assert check_arrangement('#.#.###', [1,1,3])
assert check_arrangement('.###.##....#', [3,2,1])
assert check_arrangement('....###...##......#', [3,2,1])
```

```python
def produce_possible_arrangements(springs):
    try:
        i = springs.index('?')
    except ValueError:
        i = None
    if i is None:
        return [springs]
    else:
        option_1 = springs[:i] + '.' + springs[i+1:]
        option_2 = springs[:i] + '#' + springs[i+1:]
        return produce_possible_arrangements(option_1) + produce_possible_arrangements(option_2)
```

```python
res = 0
for row in data:
    springs, numbers = row.split(' ')
    numbers = [int(x) for x in numbers.split(',')]
    res += (n_arrangements:=len([arrangement for arrangement in produce_possible_arrangements(springs) if check_arrangement(arrangement, numbers)]))
    print(f'{springs}, {numbers}: {n_arrangements}')
print(res)
```

```python
from functools import cache
```

```python
input = "input.txt"
with open(input, 'r') as infile:
    data = [l.strip() for l in infile.readlines()]
```

```python
@cache
def get_n_arrangements_for_springs(springs, numbers):
    #If there are no numbers, we return 1 if the sequence contains no '#' anymore
    #This can be an empty sequence or a sequence that contains only '.' and '?'
    if not numbers:
        return int('#' not in springs)
    #If the springs is shorter than the first number,
    #then we can't succeed anymore
    if len(springs) < numbers[0]:
        return 0
    #If we arrive here, we have at least one number and at least enough characters in the sequence
    #To match them against the first number
    #We recursively check two cases now:
    recursive_result = 0
    #a) Check if we can put the first number into the beginning of the springs
    if '.' not in springs[:numbers[0]] and (len(springs) == numbers[0] or springs[numbers[0]] != '#'):
        recursive_result += get_n_arrangements_for_springs(springs[numbers[0]+1:], numbers[1:])
    #b) Check if we can put the first number NOT into the beginning of the springs
    if springs[0] != '#':
        recursive_result += get_n_arrangements_for_springs(springs[1:], numbers)
    return recursive_result
```

```python
res = 0
for row in data:
    springs, numbers = row.split(' ')
    numbers = tuple(int(x) for x in numbers.split(','))
    springs_spread = '?'.join([springs]*5)
    numbers_spread = numbers * 5
    res += get_n_arrangements_for_springs(springs_spread, numbers_spread)
print(res)
```
