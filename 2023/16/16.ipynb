Be aware of cycles! So check if a beam already passed in there in that direction.  
Create copy of data, containing strings if light already passed through in that direction. This can later be used to count the energized tiles.

```python
def update_position(start_i, start_j, direction):
    if direction == 'N':
        return start_i-1, start_j, 'N'
    elif direction == 'E':
        return start_i, start_j+1, 'E'
    elif direction == 'S':
        return start_i+1, start_j, 'S'
    elif direction == 'W':
        return start_i, start_j-1, 'W'
```

```python
#This method just returns putative new coordinates and the direction of movement
#It does not check whether these coordinates lie in the field or whether the new position has already been visited
def move_beam(start_i,start_j,start_direction):
    current_tile = data[start_i, start_j]
    if current_tile == '.':
        return [update_position(start_i, start_j, start_direction)]
    elif current_tile == '\\':
        if direction == 'N':
            return [update_position(start_i, start_j, 'W')]
        elif direction == 'E':
            return [update_position(start_i, start_j, 'S')]
        elif direction == 'S':
            return [update_position(start_i, start_j, 'E')]
        elif direction == 'W':
            return [update_position(start_i, start_j, 'N')]
    elif current_tile == '/':
        if direction == 'N':
            return [update_position(start_i, start_j, 'E')]
        elif direction == 'E':
            return [update_position(start_i, start_j, 'N')]
        elif direction == 'S':
            return [update_position(start_i, start_j, 'W')]
        elif direction == 'W':
            return [update_position(start_i, start_j, 'S')]
    elif current_tile == '|':
        if direction in 'SN':
            return [update_position(start_i, start_j, start_direction)]
        else:
            return [update_position(start_i, start_j, 'S'), update_position(start_i, start_j, 'N')]
    elif current_tile == '-':
        if direction in 'EW':
            return [update_position(start_i, start_j, start_direction)]
        else:
            return [update_position(start_i, start_j, 'E'), update_position(start_i, start_j, 'W')]
```

```python
beam_paths = np.zeros_like(data, dtype='<U4')
```

```python
start_i, start_j, start_direction = 0, 0, 'E'
beams = [(start_i, start_j, start_direction)]
while beams:
    new_beams = []
    for (i,j,direction) in beams:
        #If this tile has been visited from this direction, no need to continue with this beam
        if direction not in beam_paths[i,j]:
            beam_paths[i,j] += direction
            #It could be that the beam is split in two now...
            moved_beams = move_beam(i,j,direction)
            #...so we treat it as a list
            for (new_i, new_j, new_direction) in moved_beams:
                if 0 <= new_i < len(data) and 0 <= new_j < len(data[0]):
                    new_beams.append((new_i, new_j, new_direction))
    beams = new_beams
```

PART II:

Find all possible start directions, then just run the above code for every one and get the maximum (the puzzle asks for the argmax but you don't actually need that).

```python
input = "input.txt"
with open(input, 'r') as infile:
    #Today we split into characters
    data = np.array([[c for c in l.strip()] for l in infile.readlines()])
```

```python
starting_configurations = \
[(i,0,'E') for i in range(len(data))] + \
[(i,len(data[0])-1,'W') for i in range(len(data))] + \
[(0,j,'S') for j in range(len(data[0]))] + \
[(len(data)-1,j,'N') for j in range(len(data[0]))]
```

```python
res = 0
for (start_i, start_j, start_direction) in starting_configurations:
    beam_paths = np.zeros_like(data, dtype='<U4')
    beams = [(start_i, start_j, start_direction)]
    while beams:
        new_beams = []
        for (i,j,direction) in beams:
            #If this tile has been visited from this direction, no need to continue with this beam
            if direction not in beam_paths[i,j]:
                beam_paths[i,j] += direction
                #It could be that the beam is split in two now...
                moved_beams = move_beam(i,j,direction)
                #...so we treat it as a list
                for (new_i, new_j, new_direction) in moved_beams:
                    if 0 <= new_i < len(data) and 0 <= new_j < len(data[0]):
                        new_beams.append((new_i, new_j, new_direction))
        beams = new_beams
    res = max(res, (beam_paths != '').sum())
```
